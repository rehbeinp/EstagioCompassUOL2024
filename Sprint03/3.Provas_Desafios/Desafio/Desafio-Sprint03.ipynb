### Gráfico contendo os 5 apps por número de instalação:

```python
import matplotlib.pyplot as plt
import pandas as pd
dados=pd.read_csv('./googleplaystore.csv',encoding='utf-8', delimiter=',')
dados_sem_repeticao=dados.drop_duplicates()
dados_ordenados=dados_sem_repeticao.sort_values(by='Installs', ascending=False)
top_cinco = dados_ordenados[["App", "Installs"]].head()
app=top_cinco["App"]
inst=top_cinco["Installs"]
instal=[]
for n in inst:
    str(n)
    n=n.strip("+")
    n=n.replace(",","")
    n=float(n)
    instal.append(n)
app=list(app)
plt.xticks(instal)
plt.barh(app,instal,  color="blue")
barWidth=0.10
plt.ylabel('Aplicativos')
plt.xlabel("Número de instalaçõs(milhões)")
plt.title("Cinco aplicativos com mais instalações")
barhWidth=0.25
```

### Gráfico pie das categorias de apps existentes ordenados de acordo com a frequência que aparecem:

```python
import matplotlib.pyplot as plt
import pandas as pd
dados=pd.read_csv('./googleplaystore.csv',encoding='utf-8', delimiter=',')
dados_sem_repeticao=dados.drop_duplicates()

catego=dados_sem_repeticao[["Category",]]
catego=catego.value_counts(normalize=True)
freq=catego
catego = catego.reset_index(level='Category')
cat=list(catego['Category'])
freq=list(freq)
labels=cat
sizes=freq
plt.rcParams['xtick.labelsize'] = 5
fg1, ax1=plt.subplots()
ax1.pie(sizes, labels=labels,startangle=5, shadow=True, radius=10.5)
ax1.axis("equal")
plt.show()
```

### Calculo 2(Forma de lista): Ordenação do Top 10 Genres por tamanho decrescente:

```python
import pandas as pd
dados=pd.read_csv('./googleplaystore.csv',encoding='utf-8', delimiter=',')
dados_sem_repeticao=dados.drop_duplicates()
generos=dados_sem_repeticao[["Genres"]]
generos=generos.value_counts()

generos=generos.sort_values(ascending=False).head(10)
generos
```

#### Gráfico: 

```python
import pandas as pd
dados=pd.read_csv('./googleplaystore.csv',encoding='utf-8', delimiter=',')
dados_sem_repeticao=dados.drop_duplicates()
generos=dados_sem_repeticao[["Genres"]]
generos=generos.value_counts()

generos=generos.sort_values(ascending=False).head(10)
generos = generos.reset_index(level='Genres')
generos
plt.rcParams['xtick.labelsize'] = 8
generos.plot( 'Genres' , 'count',title = "Quantidade de aplicativos por gênero (10 mais frequentes)" )
```
